```python
# creating film region variable
dat_clean['region'] = 'Other'
dat_clean.loc[dat_clean['country'].isin(['USA', 'Canada']), 'region'] = 'North_America'
dat_clean.loc[dat_clean['country'].isin(['UK']), 'region'] = 'UK'
dat_clean.loc[dat_clean['country'].isin(['France', 'Germany','Italy','Denmark','Ireland']), 'region'] = 'Europe'
dat_clean.loc[dat_clean['country'].isin(['Japan','Hong Kong','China','South Korea']), 'region'] = 'East_Asia'
```

```python
### ### creating dummy for regions
dummies = pd.get_dummies(dat_clean['region'], prefix = 'region', drop_first = True)
dat_clean = pd.concat([dat_clean, dummies], axis=1)
```

```python
### creating a binary variable for whether gross >= budget
dat_clean['budget_profit'] = dat_clean['gross']/ dat_clean['budget']
```

```python
### creating a binary variable for whether gross >= budget 
dat_clean['profited'] = np.where(dat_clean['budget_profit'] > 1, 1, 0)
dat_clean[['logbudget','loggross','budget_profit','profited']].head(10)
```

```python
### creat the director movie count variable
dat_clean['director_count'] = dat_clean["director_name"].map(dat_clean["director_name"].value_counts())
dat_clean['director_count'].describe()
```

```python
### calculating leave-one-out mean IMDB score for directors
# For each movie, the director's mean IMDB score EXCLUDES that movie's own score.
# This avoids endogeneity: the target-adjacent info (imdb_score) from the
# current observation no longer leaks into its own predictor.
director_groups = dat_clean.groupby("director_name")["imdb_score"]
director_sum = dat_clean["director_name"].map(director_groups.sum())
director_n = dat_clean["director_name"].map(director_groups.count())

# Leave-one-out: (sum - current_score) / (n - 1)
# For directors with only 1 movie, fall back to the global mean
global_mean = dat_clean['imdb_score'].mean()
dat_clean['director_meanimdb'] = np.where(
    director_n > 1,
    (director_sum - dat_clean['imdb_score']) / (director_n - 1),
    global_mean
)
```

```python
## LASSO regression for IMDB score prediction
predictors = ['director_count', 'loggross', 'director_facebook_likes',
              'cast_total_facebook_likes', 'genre_Drama', 'genre_Adventure',
              'genre_Biography', 'num_critic_for_reviews', 'is_english_film',
              'is_colored']

y = dat_clean['imdb_score']
X = dat_clean[predictors]

# Split into train and test FIRST, before any scaling
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1234)

# Scale using training data only to prevent data leakage
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Train set: {X_train.shape[0]} samples, Test set: {X_test.shape[0]} samples")
```

```python
### checking multicollinearity
dat_clean[predictors].corr()
```

```python
## Use LassoCV to find the best alpha via 5-fold CV on the training set only
lasso_cv = LassoCV(
    alphas=np.logspace(-6, 2, 200),
    cv=5, max_iter=2000000, random_state=1234
)
lasso_cv.fit(X_train_scaled, y_train)

best_alpha = lasso_cv.alpha_
print(f'Best alpha from cross-validation: {best_alpha:.6e}')

# Plot mean CV MSE across alphas
mean_mse = np.mean(lasso_cv.mse_path_, axis=1)
plt.figure(figsize=(8, 4))
plt.semilogx(lasso_cv.alphas_, mean_mse)
plt.axvline(best_alpha, color="red", linestyle="--", label=f"Best α={best_alpha:.2e}")
plt.xlabel("Alpha (log scale)")
plt.ylabel("Mean CV MSE")
plt.title("LASSO Cross-Validation (on training set)")
plt.legend()
plt.savefig('images/lasso_cv_mse.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
### Feature importance from LASSO (fitted on training data only)
plt.figure(figsize=(10, 6))
plt.bar(X.columns, lasso_cv.coef_)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Lasso Coefficient (standardised)")
plt.title(f"LASSO Feature Importance (α={best_alpha:.1e})")
plt.tight_layout()
plt.savefig('images/lasso_feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()

importance = pd.Series(np.abs(lasso_cv.coef_), index=X.columns)
print(importance.sort_values(ascending=False))
```

```python
### Evaluate LASSO model on the held-out test set
y_pred_train = lasso_cv.predict(X_train_scaled)
y_pred_test = lasso_cv.predict(X_test_scaled)

train_mse = mean_squared_error(y_train, y_pred_train)
test_mse = mean_squared_error(y_test, y_pred_test)
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

print(f"Train MSE: {train_mse:.4f},  Train R²: {train_r2:.4f}")
print(f"Test  MSE: {test_mse:.4f},  Test  R²: {test_r2:.4f}")

# Selected features (non-zero coefficients)
selected = pd.Series(lasso_cv.coef_, index=X.columns)
print(f"\nSelected features ({(selected != 0).sum()} of {len(selected)}):")
print(selected[selected != 0].sort_values(ascending=False))
```

```python

# Re-run the full pipeline to get updated results
import pandas as pd
import numpy as np
from sklearn.linear_model import LassoCV, LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, classification_report
from sklearn.tree import DecisionTreeClassifier

dat = pd.read_csv('movie_metadata.csv')
dat.drop_duplicates(inplace=True)
dat['logbudget'] = np.log(dat['budget'] + 1)
dat['loggross'] = np.log(dat['gross'] + 1)
dat_clean = dat.dropna()

# Feature engineering
dat_clean['is_colored'] = np.where(dat_clean['color'] == 'Color', 1, 0)
dat_clean['main_genre'] = dat_clean.genres.str.split('|').str[0]
dat_clean['is_english_film'] = np.where(dat_clean['language'] == 'English', 1, 0)
dat_clean['region'] = 'Other'
dat_clean.loc[dat_clean['country'].isin(['USA', 'Canada']), 'region'] = 'North_America'
dat_clean.loc[dat_clean['country'].isin(['UK']), 'region'] = 'UK'
dat_clean.loc[dat_clean['country'].isin(['France', 'Germany','Italy','Denmark','Ireland']), 'region'] = 'Europe'
dat_clean.loc[dat_clean['country'].isin(['Japan','Hong Kong','China','South Korea']), 'region'] = 'East_Asia'
dat_clean = dat_clean.drop(columns=['plot_keywords', 'movie_imdb_link'])
dummies = pd.get_dummies(dat_clean['main_genre'], prefix='genre', drop_first=False)
dat_clean = pd.concat([dat_clean, dummies], axis=1)
dummies = pd.get_dummies(dat_clean['region'], prefix='region', drop_first=True)
dat_clean = pd.concat([dat_clean, dummies], axis=1)
dat_clean['budget_profit'] = dat_clean['gross'] / dat_clean['budget']
dat_clean['profited'] = np.where(dat_clean['budget_profit'] > 1, 1, 0)
dat_clean['director_count'] = dat_clean["director_name"].map(dat_clean["director_name"].value_counts())

# Leave-one-out director mean IMDB
director_groups = dat_clean.groupby("director_name")["imdb_score"]
director_sum = dat_clean["director_name"].map(director_groups.sum())
director_n = dat_clean["director_name"].map(director_groups.count())
global_mean = dat_clean['imdb_score'].mean()
dat_clean['director_meanimdb'] = np.where(
    director_n > 1,
    (director_sum - dat_clean['imdb_score']) / (director_n - 1),
    global_mean
)

print("=== LASSO REGRESSION ===")
predictors = ['director_count', 'loggross', 'director_facebook_likes',
              'cast_total_facebook_likes', 'genre_Drama', 'genre_Adventure',
              'genre_Biography', 'num_critic_for_reviews', 'is_english_film',
              'is_colored']
y = dat_clean['imdb_score']
X = dat_clean[predictors]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

lasso_cv = LassoCV(alphas=np.logspace(-6, 2, 200), cv=5, max_iter=2000000, random_state=1234)
lasso_cv.fit(X_train_scaled, y_train)
best_alpha = lasso_cv.alpha_

y_pred_train = lasso_cv.predict(X_train_scaled)
y_pred_test = lasso_cv.predict(X_test_scaled)
train_mse = mean_squared_error(y_train, y_pred_train)
test_mse = mean_squared_error(y_test, y_pred_test)
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

print(f"Best alpha: {best_alpha:.6e}")
print(f"Train MSE: {train_mse:.4f}, Train R²: {train_r2:.4f}")
print(f"Test  MSE: {test_mse:.4f}, Test  R²: {test_r2:.4f}")

selected = pd.Series(lasso_cv.coef_, index=X.columns)
print(f"\nSelected features ({(selected != 0).sum()} of {len(selected)}):")
print(selected[selected != 0].sort_values(ascending=False))

print("\n=== LOGISTIC REGRESSION ===")
y2 = dat_clean['profited']
X2 = dat_clean[['director_count', 'director_meanimdb', 'region_North_America',
                'director_facebook_likes', 'cast_total_facebook_likes',
                'num_critic_for_reviews', 'num_user_for_reviews',
                'genre_Drama', 'is_english_film']]
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.25, random_state=1234)

logreg = LogisticRegression(solver='newton-cg')
cv = KFold(n_splits=5, random_state=1234, shuffle=True)
logreg_scores = cross_val_score(logreg, X2_train, y2_train, scoring="f1", cv=cv)
print(f"5-fold CV mean F1: {logreg_scores.mean():.3f}")

logreg.fit(X2_train, y2_train)
y2_pred = logreg.predict(X2_test)
print(classification_report(y2_test, y2_pred))

print("=== DECISION TREE ===")
dt = DecisionTreeClassifier(max_depth=4)
dt.fit(X2_train, y2_train)
y2_pred_dt = dt.predict(X2_test)
print(classification_report(y2_test, y2_pred_dt))
```
